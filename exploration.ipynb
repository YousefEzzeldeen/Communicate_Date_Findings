### 1- Quantitative - Quantitative Corelation Using Scatter Plot : 

```python
# BorrowerAPR and EstimatedEffectiveYield 
plt.figure(figsize = [20,6])
sb.regplot(data = df, x = 'BorrowerAPR' , y = 'EstimatedEffectiveYield',);
```

We notice that the Estimated Effective Yield is positively corelatted to the Borrower APR

### 2- Quantitative - Qualititative Corelation Using Violin and Box Plot : 

```python
# Borrower state Status and Return:
plt.figure(figsize = [22, 6])
base_color = sb.color_palette()[0]
# LEFT plot: violin plot
plt.subplot(1, 2, 1)
#Let's return the axes object
ax1 = sb.violinplot(data=df, x = 'BorrowerState' , y = 'EstimatedReturn', color=base_color, innner='quartile')
plt.xticks(rotation=90);

# RIGHT plot: box plot
plt.subplot(1, 2, 2)
sb.boxplot(data = df, x = 'BorrowerState' , y = 'EstimatedReturn',)
plt.xticks(rotation = 90)
plt.ylim(-.1,.3) ;
```

It seems that the return of most of loans is around .1 with some outlaiers

### 3- Qualititative - Qualititative Corelation Using Clusterred barcharts : 

```python
# Borrower state Status and Borrower employment status:
plt.figure(figsize = [20,6])
sb.countplot(data = df, x = 'BorrowerState', hue = 'EmploymentStatus');
```

As we see, the most employed borrowers are from CA and the most employment status in the same state is employed
