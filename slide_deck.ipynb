```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('LoanData.csv')
df.head()
```

## (Employment Status)

> The first thing we need to know is the borrower employmen status. We expect to see a variation of employment status and we'll see how this affect the other variables.

```python
# Chosing a color :
base_col = sb.color_palette()[0]
# Plotting 'Loan Status' as it is a Categorical Vriable, we use countplot in seaborn.
# code :
plt.figure(figsize = [20,6])
plt.xticks(rotation = 30)
sb.countplot(data = df , x = 'EmploymentStatus', color = base_col);
```

## (Employment Status per State)

> We now want to knoe which state has the most employees that have a loan.
we expect that the distribution is slightly different between states. Let's have a look. 

```python
# Borrower state Status and Borrower employment status:
plt.figure(figsize = [20,6])
sb.countplot(data = df, x = 'BorrowerState', hue = 'EmploymentStatus');
```

## (How does the Employment status affect the relationbetween BorrowerAPR and EstimatedEffectiveYield )



```python
#Faceting in Multi variable : 
bin_edges = np.arange(12, 58+2, 2)
g = sb.FacetGrid(data = df, col = 'EmploymentStatus', col_wrap=3, sharey=False, margin_titles = True)
g.map(plt.scatter,'BorrowerAPR','EstimatedEffectiveYield');
```
